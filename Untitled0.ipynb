```python
from transformers import GPT2LMHeadModel, GPT2Tokenizer

# Load the tokenizer and model
tokenizer = GPT2Tokenizer.from_pretrained("gpt2")
model = GPT2LMHeadModel.from_pretrained("gpt2")
```

```python
prompt = "The secret to happiness is hidden in..."
```

```python
input_ids = tokenizer.encode(prompt, return_tensors="pt")

# Generate output
output = model.generate(
    input_ids,
    max_length=100,
    num_return_sequences=1,
    no_repeat_ngram_size=2,
    do_sample=True,
    top_k=50,
    top_p=0.95,
    temperature=0.8
)

# Decode and print the result
generated_text = tokenizer.decode(output[0], skip_special_tokens=True)
print(generated_text)
```

```python
# Step 1: Install Hugging Face Transformers
!pip install transformers

# Step 2: Import libraries
from transformers import pipeline

# Step 3: Load GPT-2 pipeline
generator = pipeline("text-generation", model="gpt2")

# Step 4: Function to get capital using text generation
def get_capital_gpt2(country):
    prompt = f"The capital of {country} is"
    output = generator(prompt, max_length=20, num_return_sequences=1)[0]['generated_text']
    print(output)

# Step 5: Try it!
get_capital_gpt2("France")
get_capital_gpt2("Japan")
get_capital_gpt2("India")
```

```python
!pip install transformers

# Step 2: Import the pipeline
from transformers import pipeline

# Step 3: Create the question-answering pipeline
qa_pipeline = pipeline("question-answering")

# Step 4: Define context and question
context = """
The capital of France is Paris. It is known for the Eiffel Tower, art museums like the Louvre, and its rich cultural history.
"""

question = "What is the capital of France?"

# Step 5: Ask the question
result = qa_pipeline(question=question, context=context)

# Step 6: Print the result
print(f"Answer: {result['answer']}")
```

```python
# Step 1: Install required libraries
!pip install transformers gradio

# Step 2: Import libraries
from transformers import pipeline
import gradio as gr

# Step 3: Load the QA pipeline
qa_pipeline = pipeline("question-answering",)

# Step 4: Define function for Gradio interface
def answer_question(context, question):
    result = qa_pipeline(question=question, context=context)
    return result["answer"]

# Step 5: Create the Gradio interface
iface = gr.Interface(
    fn=answer_question,
    inputs=[
        gr.Textbox(label="Context", lines=5, placeholder="Paste your article or paragraph here..."),
        gr.Textbox(label="Question", placeholder="Ask a question based on the context")
    ],
    outputs="text",
    title="AI Question Answering App",
    description="Ask a question based on the given context using Hugging Face Transformers"
)

# Step 6: Launch the app
iface.launch()
```
